```python
num_cols = []
cat_cols = []
response_set = [response]
feature_names = df_all.columns.tolist() #- response_set #- train_test
```

```python
uniques = {}
k = 20
for i in feature_names:
    if df_all[i].dtype is df_all['CAR_MODEL'].dtype:
        cat_cols.append(i)
    elif len(df_all[i].unique()) < 20:
        cat_cols.append(i)
    else:
        num_cols.append(i)
```

```python
df_num = df_all[num_cols]
```

```python
df_cat = df_all[cat_cols]
```

```python
df_cat.head()
```

```python
df_num = df_num.drop('VIN_HASHED', axis=1)
```

```python
def replace_vals(df,column,name_prfx):
    '''
    INPUTS: 
     - df: dataframe
     - column: column that must be replaced with fake names
     - prefix eg. car_ in:(car_1, car_2,)
    
    OUTPUTS: df with values replaced in the specific column
    
    '''
    unique = df[column].unique()
    j=0
    for i in unique:
        print(i)
        df[column] = df[column].replace(i,name_prfx+str(j))
        j=j+1
```

```python
def normalizer(df_num):
    
    df_norm = (df_num - df_num.mean()) / (df_num.max() - df_num.min()) 

    #sanity check:
    assert sum(df_norm.max()-df_norm.min())  == len(df_norm.columns)

    return df_norm
```

```python
replace_vals(df_cat,'CAR_MODEL','car_')
```

```python
df_cat.head()
```

```python
df_norm = normalizer(df_num)
```

```python
df_norm.head()
```

```python
df_final = pd.concat([df_all['VIN_HASHED'],df_cat,df_norm],axis=1)
```
